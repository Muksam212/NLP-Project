Problem defination -: Our goal is to look at the transcripts of various comedians and note their similarities and differences

# Loading the data

```python
import pickle
data = {}
#key : value -> key(comedian name), value: (transcript)

comedians = ["louis", "dave", "ricky", "bo", "bill",
            "jim", "john", "hasan", "ali", "anthony", "mike", "joe"]

for i, c in enumerate(comedians):
    with open('transcripts/' + c + ".txt", "rb") as file:
        data[c] = pickle.load(file)
```

```python
data.keys()
```

```python
data['ricky'][:2]
```

```python
def combine_text(list_of_text):
    combine_text = ' '.join(list_of_text)
    return combine_text
```

```python
type(data)
```

```python
#dictonary with key as comedian name and value is their transcript
#dictionary comprehension {k: v for vars in interable}
#items will make dictionary into the list of tuples

data_combine = {key: [combine_text(value)] for (key, value) in data.items()}
```

```python
#out data dictionary converted to data from pandas
import pandas as pd

data_df = pd.DataFrame.from_dict(data_combine).transpose()
```

```python
data_df.columns = ["transcript"]
data_df = data_df.sort_index()

data_df
```

```python
data_df.transcript.loc["ali"]
```

# Cleaning the data

```python
import re #to define the patterns
import string
```

```python
import warnings
warnings.filterwarnings("ignore")
def clean_text(text):
    text = text.lower()
    text = re.sub('\[.*?\]', '', text) #hello -> remove any text with square bracket
    text = re.sub('[%s]' % re.escape(string.punctuation), '', text)
    #text containing the numbers
    #\w means words and \d means digits
    text = re.sub('\w*\d\w*', '', text)
    return text
    
```

```python
round1 = lambda x: clean_text(x)
```

```python
#look the update text
data_clean = pd.DataFrame(data_df.transcript.apply(round1))
data_clean
```

```python
#next round
def clean_text_round2(text):
    text = re.sub('[''""...]', "", text)
    text = re.sub("\n", "", text)
    text = re.sub("\xa0", "", text)
    return text

round2 =lambda x: clean_text_round2(x)
```

```python
data_clean = pd.DataFrame(data_clean.transcript.apply(round2))
data_clean.transcript.loc["ricky"]
```

# To remove stopwords

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(stop_words = "english")
data_cv = cv.fit_transform(data_clean.transcript)
```

```python
#we are going to use document-term matrix to convert text into the matrix format
data_dtm = pd.DataFrame(data_cv.toarray(), columns = cv.get_feature_names_out())
```

```python
#data clean index is an comedian name
data_dtm.index = data_clean.index
```

```python
data_dtm
```

# Exploratory Data Analysis

# Most common words

```python
#find the top 30 words said by each comedians

data = data_dtm.transpose()

top_dict = {}
for c in data.columns:
    top = data[c].sort_values(ascending = False).head(10)
    top_dict[c] = list(zip(top.index, top.values))
    
print(top_dict)
```

```python
#Print the top 15 words spoken by each comedian

for comedian, top_words in top_dict.items():
    print(comedian)
    #using the list comprehension
    print(" , ".join([word for word, count in top_words[0:14]]))
    print('==================================')
```

```python
#all the top words that has been spoken by the comedains
from collections import Counter

words = [] #making empty list to append the code later on

for comedian in data.columns:
    top = [word for word, count in top_dict[comedian]]
    for t in top:
        words.append(t)

words
```

```python
Counter(words).most_common()
```

```python
add_stop_words = [word for word, count in Counter(words).most_common() if count > 6]
```

```python
print(add_stop_words)
```

```python
from sklearn.feature_extraction import text
from sklearn.feature_extraction.text import CountVectorizer
```

```python
stop_words = text.ENGLISH_STOP_WORDS.union(add_stop_words)
```

# WordCloud 

```python
#creating an wordcloud to visualize which text has the high frequency.
from wordcloud import WordCloud
```

```python
wc = WordCloud(stopwords = stop_words, background_color = "white",
              colormap = "Dark2", max_font_size = 150,
              random_state = 42)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams["figure.figsize"] = [16, 6] #dimension of figure size

#create subplots
names = ["Ali", "Anthony", "Bill", "Bo",
        "Dave", "Hasan", "Jim", "Joe", "John", "Louis",
        "Mike", "Ricky"]


for index, comedian in enumerate(data.columns):
    wc.generate(data_clean.transcript[comedian])
    plt.subplot(3, 4, index + 1)
    plt.imshow(wc, interpolation = "bilinear")
    plt.title(names[index])
```

```python
data_bad_words = data.transpose()[['fucking','fuck','shit']]
```

```python
data_profanity = pd.concat([data_bad_words.fucking + data_bad_words.fuck, data_bad_words.shit], axis = 1)
```

```python
data_profanity.columns = ['f_word', 's_word']
```

```python
data_profanity
```

```python
#scatter plot to represent each of the comedians

for i, comedian in enumerate(data_profanity.index):
    x = data_profanity.f_word.loc[comedian]
    y = data_profanity.s_word.loc[comedian]
    
    plt.scatter(x, y, color = "blue")
    
    plt.text(x + 1.5, y + 0.5, names[i], fontsize = 10)
    
plt.xlabel("Number of F Words")
plt.ylabel("Number of S Words")

plt.show()
```

# Sentimental Analysis

```python
data_clean
```

```python
from textblob import TextBlob
```

```python
pol = lambda x: TextBlob(x).sentiment.polarity

sub = lambda x: TextBlob(x).sentiment.subjectivity
```

```python
data_clean["polarity"] = data_clean["transcript"].apply(pol)
data_clean["subjectivity"] = data_clean["transcript"].apply(sub)
```

```python
data_clean
```


# Plot the Polarity and Subjectivity

```python
names = ["Ali", "Anthony", "Bill", "Bo",
    "Dave", "Hasan", "Jim", "Joe", "John", "Louis",
    "Mike", "Ricky"]


for index, comedian in enumerate(data_clean.index):
    

    x = data_clean.polarity.loc[comedian]
    y = data_clean.subjectivity.loc[comedian]
    
    plt.scatter(x, y, color = "blue")
    
    plt.text(x + 0.01, y + 0.01, names[index])
    
    
plt.show()
    
    
```
